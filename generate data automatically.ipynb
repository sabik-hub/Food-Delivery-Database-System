```python
from faker import Faker
import random

fake = Faker()

# Enable uniqueness
fake.unique.clear()

# -------------------------
# 1. Generate CUSTOMERS
# -------------------------
customers = []

for _ in range(20):
    customer = {
        "name": fake.name(),
        "email": fake.unique.email(),
        "phone": fake.unique.phone_number(),
        "address": fake.address().replace("\n", ", ")
    }
    customers.append(customer)

print("CUSTOMERS:")
for c in customers:
    print(c)

# -------------------------
# 2. Generate RESTAURANTS
# -------------------------
restaurants = []

for _ in range(5):
    restaurant = {
        "name": fake.company(),
        "phone": fake.unique.phone_number(),
        "location": fake.city()
    }
    restaurants.append(restaurant)

print("\nRESTAURANTS:")
for r in restaurants:
    print(r)

# -------------------------
# 3. Generate FOOD ITEMS
# -------------------------
fooditems = []

food_names = ["Pizza", "Burger", "Pasta", "Biryani", "Dosa", "Sandwich"]

for i in range(10):
    food = {
        "food_name": random.choice(food_names),
        "price": round(random.uniform(50, 500), 2),
        "availability": True
    }
    fooditems.append(food)

print("\nFOOD ITEMS:")
for f in fooditems:
    print(f)

# -------------------------
# 4. Generate DELIVERY PARTNERS
# -------------------------
partners = []

for _ in range(10):
    partner = {
        "name": fake.name(),
        "phone": fake.unique.phone_number(),
        "vehicle_type": random.choice(["Bike", "Scooter", "Cycle"]),
        "availability": "available"
    }
    partners.append(partner)

print("\nDELIVERY PARTNERS:")
for p in partners:
    print(p)
```

```python
from faker import Faker
import random

fake = Faker()

restaurant_types = {

    "Cafe":("07:00:00","21:00:00"),

    "Fast Food":("10:00:00","23:00:00"),

    "Bakery":("06:00:00","20:00:00"),

    "Family Restaurant":("11:00:00","22:00:00"),

    "Biryani House":("12:00:00","23:59:00"),

    "24x7 Kitchen":("00:00:00","23:59:59")
}

for _ in range(20):

    category = random.choice(list(restaurant_types.keys()))

    opening_time, closing_time = restaurant_types[category]

    restaurant_name = (
        fake.company() + " " + category
    )

    location = fake.city()

    phone = fake.unique.phone_number()

    rating = round(
        random.uniform(3.5,5.0),
        1
    )

    cur.execute("""
    INSERT INTO restaurants
    (
        restaurant_name,
        location,
        phone,
        rating,
        opening_time,
        closing_time
    )

    VALUES
    (%s,%s,%s,%s,%s,%s)
    """,

    (
        restaurant_name,
        location,
        phone,
        rating,
        opening_time,
        closing_time
    ))

conn.commit()

print("Restaurants inserted")
```

```python
import random

foods = {
    "Fast Food": ["Burger","Pizza","Sandwich","French Fries"],
    "Indian": ["Biryani","Dosa","Butter Chicken","Parotta"],
    "Chinese": ["Noodles","Fried Rice","Manchurian"],
    "Dessert": ["Ice Cream","Cake","Brownie"],
    "Beverage": ["Coffee","Tea","Juice"]
}

restaurant_ids = list(range(2001,2021))

for _ in range(100):

    category = random.choice(list(foods.keys()))

    food_name = random.choice(
        foods[category]
    )

    cur.execute("""
        INSERT INTO fooditems
        (restaurant_id,
         food_name,
         category,
         price,
         availability)

        VALUES(%s,%s,%s,%s,%s)
    """,

    (
        random.choice(restaurant_ids),
        food_name,
        category,
        round(random.uniform(50,500),2),
        random.choice([True,True,True,False])
    ))

conn.commit()
```

```python
import random

statuses = [
    'placed',
    'preparing',
    'out_for_delivery',
    'delivered',
    'cancelled'
]

for _ in range(500):

    customer_id = random.randint(1001,1100)

    partner_id = random.randint(4001,4030)

    order_date = fake.date_time_this_year()

    total_amount = round(
        random.uniform(150,1200),2
    )

    status = random.choice(statuses)

    cur.execute("""
        INSERT INTO orders
        (
            customer_id,
            partner_id,
            order_date,
            total_amount,
            status
        )
        VALUES(%s,%s,%s,%s,%s)
    """,

    (
        customer_id,
        partner_id,
        order_date,
        total_amount,
        status
    ))

conn.commit()
```

```python
methods = [
    'UPI',
    'Card',
    'Cash'
]

statuses = [
    'completed',
    'pending',
    'failed',
    'refunded'
]

for order_id in range(100001,100501):

    amount = round(
        random.uniform(150,1200),2
    )

    payment_date = fake.date_time_this_year()

    cur.execute("""
        INSERT INTO payments
        (
            order_id,
            amount,
            payment_method,
            payment_status,
            payment_date
        )
        VALUES(%s,%s,%s,%s,%s)
    """,

    (
        order_id,
        amount,
        random.choice(methods),
        random.choice(statuses),
        payment_date
    ))

conn.commit()
```
